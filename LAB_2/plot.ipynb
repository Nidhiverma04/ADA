```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
size = [3, 6, 12, 24, 48, 96, 192, 384, 768, 1536, 3072, 6144, 12288, 24576]
time = [0, 0.00012, 0, 0, 0, 0, 4e-005, 0.00021, 0.00051, 0.00202, 0.00715, 0.02758, 0.11625, 0.44184]
i = 8
ref = size[i]
n_squared = [size**2 * time[i] / (ref**2) if size != 0 else 0 for size in size]
 
plt.grid(True)

plt.plot(size, time, marker='o', label='Selection Sort Time')
plt.plot(size, n_squared, marker='x', linestyle='--', label='O(n^2) Complexity')

plt.xlabel("Array Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Selection Sort Time Complexity")
plt.show()
```

```python
size = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384]
time = [0, 0, 0, 0, 0, 0, 0.00016, 0.00015, 0.00088, 0.00363, 0.01333, 0.04935, 0.19876, 0.79614]
ref = size[8]
n_squared = [size**2 * time[8] / (ref**2) if size != 0 else 0 for size in size]
 
plt.grid(True)

plt.plot(size, time, marker='o', label='Bubble Sort Time')
plt.plot(size, n_squared, marker='x', linestyle='--', label='O(n^2) Complexity')

plt.xlabel("Array Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Bubble Sort Time Complexity")
plt.show()
```

```python
size = [3, 5, 9, 17, 33, 65, 129, 257, 513, 1025, 2049, 4097, 8593, 16385]
time = [0, 0, 0, 0, 0, 0, 0, 2903, 0, 9891, 67967, 279560, 1508706, 9265309]


reference_size = size[1]
reference_time = time[1] if time[1] != 0 else 1  

n_squared = [(size**2 * reference_time) / (reference_size**2) for size in size]

plt.grid(True)

plt.plot(size, time, marker='o', label='Magic Square Time')
plt.plot(size, n_squared, marker='x', linestyle='--', label='O(n^2) Complexity')

plt.xlabel(f"Array Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Magic Square Time Complexity")
plt.show()
```
