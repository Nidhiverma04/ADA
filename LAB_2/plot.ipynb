```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

size = np.array(pd.read_csv('size.csv', header=None)).flatten()
time = np.array(pd.read_csv('time.csv', header=None)).flatten()

i = 4
ref = size[i]
n_squared = [s * np.log2(s) * time[i] / (ref * np.log2(ref)) for s in size]

plt.grid(True)
plt.plot(size, time, marker='o', label='Heap Sort Time')
plt.plot(size, n_squared, marker='x', linestyle='--', label='O(n log n) Complexity')

plt.xlabel("Array Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Heap Sort Time Complexity")
plt.show()
```

```python
import matplotlib.pyplot as plt

size = [10, 20, 40, 80, 160, 320, 640, 1280, 2560, 5120, 10240, 20480, 40960]
time = [2.5e-005, 4.5e-005, 0.000105, 0.000215, 0.00035, 0.00077, 0.00172, 0.003805, 0.008045, 0.021615, 0.055605, 0.130295, 0.243555]

ref = 12
reference_size = size[ref]
reference_time = time[ref] if time[ref] != 0 else 1  

n_linear = [(s * reference_time) / reference_size for s in size]

plt.grid(True)
plt.plot(size, time, marker='o', label='Cosine Similarity Time')
plt.plot(size, n_linear, marker='x', linestyle='--', label='O(n) Complexity')
plt.xlabel("Array Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Cosine Similarity Time Complexity")
plt.show()
```
