```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

size = np.array(pd.read_csv('size.csv', header=None)).flatten()
time = np.array(pd.read_csv('time.csv', header=None)).flatten()

i = 0
ref = size[i]
n_squared = (np.log2(size) / np.log2(ref)) * time[i]

plt.grid(True)
plt.plot(size, time, marker='o', label='Peak 1D Time')
plt.plot(size, n_squared, marker='x', linestyle='--', label='O(log n) Complexity')

plt.xscale('log')
if np.all(time > 0):  
    plt.yscale('log') 
else:
    plt.yscale('symlog')  

plt.xlabel("Array Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Peak 1D Time Complexity")
plt.show()
```
