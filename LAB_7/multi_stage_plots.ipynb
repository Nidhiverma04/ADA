```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_mst(edges, n):
    G = nx.Graph()
    G.add_weighted_edges_from(edges)


    mst = nx.minimum_spanning_tree(G, algorithm="kruskal")

    pos = nx.spring_layout(G)  
    plt.figure(figsize=(8, 6))
    plt.title("Original graph")
   
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10)

    # nx.draw(mst, pos, with_labels=True, node_color='lightgreen', node_size=500, font_size=10, edge_color='red', width=3)
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): w for u, v, w in edges})

    plt.title("Multistage Graph Plot")
   

edges = [
    (0, 2, 3),
    (0, 3, 5),
    (1, 2, 5),
    (1, 3, 6),
    (2, 3, 5)
]
n = 4 

plot_mst(edges, n)
```

Forward substitution

```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_mst(edges, n, edges_msg):
    G = nx.Graph()
    G.add_weighted_edges_from(edges)
    G1 = nx.Graph()
    G1.add_weighted_edges_from(edges_msg)

    pos = nx.spring_layout(G)  

    plt.figure(figsize=(8, 6))
    plt.title("Original graph")
   
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10)

    nx.draw(G1, pos, with_labels=True, node_color='lightgreen', node_size=500, font_size=10, edge_color='red', width=3)
    nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): w for u, v, w in edges})

    plt.title("Multistage Graph Forward Substitution")
   

edges = [
    (0, 1, 2),
    (0, 2, 3),
    (1, 3, 4),
    (1, 4, 1),
    (2, 3, 1),
    (2, 4, 5),    (3, 4, 2)
]

edges_msg = [
    (0, 1, 2),
    (1, 4, 1)
]
n = 5 

plot_mst(edges, n, edges_msg)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

size = np.array(pd.read_csv('size.csv', header=None)).flatten()
time = np.array(pd.read_csv('time.csv', header=None)).flatten()


i = 9
scale_factor = time[i] / (size[i] ** 2)
n_log_n = (size ** 2) * scale_factor
idx = 0
plt.plot(size[idx:], time[idx:], 'o-', label="No. of Stages Time")
plt.plot(size[idx:], n_log_n[idx:], 'x--', label='O(n^2) Scaled')
plt.title("No. of Stages in MSG Plot")
plt.xlabel('No. of vertices'); plt.ylabel('Time taken (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
